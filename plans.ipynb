## Даша бот

Сделать бота, который будет автоматически по пользователям из списка проверять их нахождение в чате. Если есть какой-то пользователь, который не состоит в списке, то его нужно кикать из чата/канала

## План работы

1. Оформить базу данных для хранения списка пользователей и чата
2. Создать бота в телеграме, получить токен
3. Написать команды, которые будут добавлять и удалять пользователей из списка
4. Написать скрипт, который будет проверять наличие пользователей в чате и кикать пользователей, которых нет в списке
5. В чате боту можно делать запросы (если пользователь, который отправляет запросы - администратор), на которые он будет отвечать информацией о количестве пользователей в чате и количестве пользователей в списке, наличии конкретного пользователя в вайт листе и т.п.


## Стуктура проекта

1. `bot.py` - основной файл, в котором находится бот
2. `whitelist.db` - база данных, в которой хранится список пользователей и чатов
3. `requirements.txt` - файл с зависимостями
4. `whitelist.py` - файл с функциями для работы с базой данных
5. `README.md` - описание проекта
